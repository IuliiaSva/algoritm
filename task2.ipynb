```python
def binary_search (arr,a):
    low=0
    high=len(arr)-1
    while low<=high:
        mid=(low+high)//2
        guess=arr[mid]
        if guess==a:
            return mid
        if guess>a:
            high=mid-1
        else:
            low=mid+1
    return -1
arr = [1, 2, 6, 8, 3, 4, 5]
arr.sort ()
binary_search (arr,3)
```

```python
import timeit
def binary_search (arr,a):
    low=0
    high=len(arr)-1
    while low<=high:
        mid=(low+high)//2
        guess=arr[mid]
        if guess==a:
            return mid
        if guess>a:
            high=mid-1
        else:
            low=mid+1
    return -1
s = """def test_binary_search():
    arr = [1, 2, 6, 8, 3, 4, 5]
    arr.sort()
    binary_search(arr,6)
    """
print(timeit.timeit(stmt=s, number=1000))
# Пространственая сложность O(log n), время выполнения от 3 до 6, не требуется просматривать каждый объект.
```

```python
import random
import time
import matplotlib.pyplot as plt
def binary_search (arr,a):
    low=0
    high=len(arr)-1
    while low<=high:
        mid=(low+high)//2
        guess=arr[mid]
        if guess==a:
            return mid
        if guess>a:
            high=mid-1
        else:
            low=mid+1
    return -1
def plot_search_time(sizes, iterations=100):
    timez = []
    for size in sizes:
        total_time = 0
        for _ in range(iterations):
            my_list = [random.randint(0, size) for _ in range(size)]
            my_list.sort()
            target = random.randint(0, size)
            star_time = time.time()
            binary_search (my_list, target)
            end_time = time.time()
            total_time += end_time - star_time
        timez.append(total_time)
    plt.plot(sizes, timez, marker='o')
    plt.title('Время выполения')
    plt.xlabel('Размер списка')
    plt.ylabel('Время, секунды')
    plt.grid(True)
    plt.show()
if __name__ == '__main__':
    sizes = [100, 200, 300, 400, 500]
    plot_search_time(sizes)

def index (arr,a):
    for i in range (len(arr)):
     if arr[i]==a:
      return i
    return -1
def plot_search_time (sizes, iterations=100):
    timez = []
    for size in sizes:
        total_time = 0
        for _ in range(iterations):
            my_list= [random.randint(0,size) for _ in range(size)]
            target = random.randint (0,size)
            star_time = time.time ()
            index (my_list,target)
            end_time = time.time ()
            total_time += end_time - star_time
        timez.append(total_time)
    plt.plot (sizes, timez, marker='o')
    plt.title ('Время выполения')
    plt.xlabel ('Размер списка')
    plt.ylabel ('Время, секунды')
    plt.grid (True)
    plt.show()
if __name__ == '__main__':
    sizes = [100,200,300,400,500]
    plot_search_time(sizes)

#Бинарный поиск требует больше времени на саму операцию, но меньше количество операций, так как не требуется проходить каждый элемент, при большом массиве он эффективен. При небольшом списке эффективнее линейный поиск, так как просмотреть каждый элекмент становится быстрее, без совершения дополнительных операций.
```
